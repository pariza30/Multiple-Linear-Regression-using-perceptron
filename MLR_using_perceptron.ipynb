```python
import numpy as np
import pandas as pd
df=pd.read_csv("/content/multiple_linear_regression_dataset 2.csv")

```

```python
x=df[["age","experience"]].values
y=df["income"].values
x = (x - x.mean(axis=0)) / x.std(axis=0)
```

```python
bias=np.ones((x.shape[0],1))
```

```python
x=np.hstack((x,bias))
```

```python
np.random.seed(42)
```

```python
weights=np.random.randn(x.shape[1])
```

```python
weights
```

```python
lr=0.0001
epochs=2000
len(x)
```

```python
y=df["income"].values
```

```python
for epoch in range(epochs):
  total_error=0
  for i in range(len(x)):
    y_pred=np.dot(x[i],weights)
    err=y[i]-y_pred
    total_error+=err**2
    weights=weights+lr*err*x[i]
  if(epoch%200==0):
    print(f"epoch{epoch}|mse={total_error/len(x):.4f}")
```

```python
weights
```

```python
for i in range(5):
    pred = np.dot(x[i], weights)
    print(f"Actual: {y[i]}; Predicted: {pred:.2f}")
```
